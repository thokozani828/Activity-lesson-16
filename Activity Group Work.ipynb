```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph object
G = nx.Graph()

# Add edges to the graph with weights
G.add_edge('Durban', 'Pietermaritzburg', weight=89)
G.add_edge('Durban', 'Richards Bay', weight=112)
G.add_edge('Pietermaritzburg', 'Richards Bay', weight=70)
G.add_edge('Pietermaritzburg', 'Harrismith', weight=209)
G.add_edge('Harrismith', 'Richards Bay', weight=100)
G.add_edge('Harrismith', 'Volksrust', weight=41)
G.add_edge('Harrismith', 'Johannesburg', weight=210)
G.add_edge('Richards Bay', 'Volksrust', weight=106)
G.add_edge('Volksrust', 'Johannesburg', weight=106)

# Function to get the shortest path and its cost
def find_shortest_path(graph, start, end):
    path = nx.shortest_path(graph, source=start, target=end, weight='weight')
    cost = nx.shortest_path_length(graph, source=start, target=end, weight='weight')
    return path, cost

# Example usage: Find the shortest path from Durban to Johannesburg
start_city = 'Durban'
end_city = 'Johannesburg'
shortest_path, total_cost = find_shortest_path(G, start_city, end_city)

# Print the result
print(f"The shortest path from {start_city} to {end_city} is: {' -> '.join(shortest_path)} with a total cost of {total_cost} kilometres.")

# Visualize the graph
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 8))

# Draw the nodes and edges
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Highlight the shortest path
edge_colors = ['black' if not (u, v) in list(zip(shortest_path, shortest_path[1:])) else 'red' for u, v in G.edges()]
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=2)

# Show the plot
plt.title(f"Shortest Path from {start_city} to {end_city}")
plt.show()
```
